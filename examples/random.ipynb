```python
import jax.random as rnd
import jax.numpy as jnp
import matplotlib.pyplot as plt
key = rnd.key(42)
key,subkey = rnd.split(key)

r=rnd.normal(subkey,shape=(1e8,))
```

```python
plt.hist(r,bins=1000)
x=jnp.arange(-10,10,0.0001)
plt.plot(x,1.05e6/jnp.sqrt(2*jnp.pi)*jnp.exp(-x**2/2))
```
